```python
import pandas as pd 

import matplotlib
from matplotlib import pyplot as plt

import seaborn as sns
```

```python
#grab your favorite wines from the last Pandas lab

data = pd.read_csv("https://raw.githubusercontent.com/Zipcoder/DataEngineering.Labs.WineQuality/master/Part%203%20(feature%20engineering)/my%20favorite%20wines.csv?token=ALOLXOOXFEEZ45XROEBUDGC6CDBVK", index_col=0)
```

```python
data.tail()
```

Ok - let's first grab a slice of this df. We're going to do some visualization, so let's drop the last three columns for now.

```python
data.drop(labels=['color', 'sweet?', 'I like it'], axis=1, inplace=True)
```

Take note of the 'inplace' argument above. If we didn't specify it as True, we'd just get a view of the df, without actually altering it.

```python
data.tail()
```

There we go! 

Ok - now for the pretty colors. There are lots of ways to start the process of EDA, so let's start with looking at basic correlations.

```python
plt.figure(figsize=(12,10), dpi= 240)
sns.heatmap(data.corr(), xticklabels=data.corr().columns, 
            yticklabels=data.corr().columns, cmap='RdYlGn', center=0, annot=True)

# Decorations
plt.title('Correlogram of My Favorite Wine Qualities', fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=12)
plt.show()
```
